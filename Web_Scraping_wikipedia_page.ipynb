#### Import the requests library to make HTTP request

```python
import requests
```

```python
# Setting the browser agent string in the user_agent variable
user_agent ="Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/115.0.0.0 Safari/537.36"
```

```python
#Storing the Wikipedia URL in the url variable
url = "https://en.wikipedia.org/wiki/List_of_state_and_union_territory_capitals_in_India"
```

```python
#Storing the User-Agent request header in the headers variable
headers = {'User-Agent': user_agent}
```

```python
#Using the get function to get the URL content
page = requests.get(url, headers)
```

```python
# Importing BeautifulSoup from bs4 and setting the soup variable to get the page content
from bs4 import BeautifulSoup
soup = BeautifulSoup(page.content, 'html.parser')
```

```python
#Let us just have a look at how the response looks like
print(page.content)
```

#### Get all the tables from that Wikipedia page and store them in the all_tables variable. To achieve this, we will be using the find_all function of BeautifulSoup

```python
all_tables = soup.find_all('table')
```

####  selecting one particular table from all the tables on that page in which data is there, then store it in the right_table variable. We will do this using the find function on BeautifulSoup

```python
right_table = soup.find('table', class_='wikitable sortable')
```

```python
# # Printing the table we saved in the right_table variable
print(right_table)
```

#### Now defining various lists to store the data from this table

```python
A = [] # For Number
B = [] # For State/UT
C = [] # For Administrative capitals
D = [] # For Legislative capitals
E = [] # For Judiciary capitals
F = [] # For Year capital was established
G = [] # The Former capital
```

#### storing the data in the respective lists from the Wikipedia table

```python
for row in right_table.findAll("tr"):
    cells = row.findAll('td')
    states = row.findAll('th') #To store second column data

    if 0 < len(cells):
        A.append(cells[0].find(text=True).rstrip())
    if 0 < len(states):
        B.append(states[0].find(text=True).rstrip())
    if 1 < len(cells):
        C.append(cells[1].find(text=True).rstrip())
    if 2 < len(cells):
        D.append(cells[2].find(text=True).rstrip())
    if 3 < len(cells):
        E.append(cells[3].find(text=True).rstrip())
    if 4 < len(cells):
        F.append(cells[4].find(text=True).rstrip())
    if 5 < len(cells):
        G.append(cells[5].find(text=True).rstrip())
```

```python
# Now let's view the first and second lists in 2 different cells
print(A)
print(B)
```
